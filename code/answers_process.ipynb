```python
import string
import pandas as pd
```

```python
df = pd.read_json('./answers.jsonl', lines=True)
df.head()
```

```python
law_ids = dict()

for answer in df['answers'].tolist():
    answer = [x.split(' > ')[0].strip() for x in answer]
    for a in answer:
        if a not in law_ids:
            law_ids[a] = 1
        else:
            law_ids[a] += 1

law_ids = pd.DataFrame.from_dict(law_ids, orient='index', columns=['count'])
law_ids = law_ids.sort_values(by=['count'], ascending=False)
law_ids = law_ids.reset_index()
law_id = '58/2014/QH13'
new_df = df[df['answers'].apply(lambda x: any(law_id in y for y in x))]
new_df = new_df.reset_index(drop=True)
new_df.head()
```

```python
article = dict()
for answer in new_df['answers'].tolist():
    for x in answer:
        if '58/2014/QH13' not in x:
            continue
        split = x.split('>')
        if len(split) > 1:
            temp = split[1].lower().strip()
            if "điều" not in temp:
                continue
            temp = temp.translate(str.maketrans('', '', string.punctuation))
            if temp not in article:
                article[temp] = 1
            else:
                article[temp] += 1
article = pd.DataFrame.from_dict(article, orient='index', columns=['count'])
article.head()
```
